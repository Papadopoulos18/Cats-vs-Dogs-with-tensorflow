```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import TensorBoard
import pickle
import time


NAME = "Cats-vs-dogs-CNN"


# # οι επομενες 2 γραμμες δεν ειναι αναγκαιες
# gpu_options = tf.GPUOptions(per_process_gpu_memory_fraction=0.333)  #λες οτι θελω αυτο το model να "παρει" το 33.3% της GPU  
# sess = tf.Session(config=tf.ConfigProto(gpu_options=gpu_options))

pickle_in = open("X.pickle","rb")
X = pickle.load(pickle_in)

pickle_in = open("y.pickle","rb")
y = pickle.load(pickle_in)

X = X/255.0

#BUILD MY CNN
model = Sequential()

model.add(Conv2D(128, (3, 3), input_shape=X.shape[1:])) #διοτι εαν κανω print(X.shape ->(24904,1,50,50))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(256, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

#τα κανω flatten για να εινια το input sto FCNN (FULLY CONNECTED NN)
model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors


model.add(Dense(1))
model.add(Activation('sigmoid'))

tensorboard = TensorBoard(log_dir="C:\\Users\\aggelos\\sentex\\deep learning with  python, tensorflow, keras\\logs\\{}".format(NAME))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X, y, batch_size=32, epochs=10, validation_split=0.1, callbacks = [tensorboard])


model.save('CNN.model')
```
